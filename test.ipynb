# TRAIN

```python
from ultralytics import YOLO

def train_yolo11():
    model = YOLO("/content/drive/MyDrive/Car_Detection/yolo11n.pt") 
    results = model.train(
        data="/content/drive/MyDrive/Car_Detection/dataset/data.yaml", # ที่อยู่ไฟล์ yaml ของชุดข้อมูลคุณ
        epochs=100,                    # จำนวนรอบการเทรน
        imgsz=1024,                     # ขนาดภาพที่ใช้เทรน
        batch=32,                      # ปรับตามความแรงของ GPU (เช่น 16, 32, -1 คือ auto)
        device=0,                      # ลำดับของ GPU (ถ้าใช้ CPU ให้ใส่ "cpu")
        name="yolo11_custom_model"     # ชื่อโฟลเดอร์สำหรับเก็บผลลัพธ์
    )

if __name__ == "__main__":
    train_yolo11()
```
